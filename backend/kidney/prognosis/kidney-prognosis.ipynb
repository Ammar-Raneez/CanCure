```python
# read dataset to pandas dataframe
import io

df = pd.read_csv(io.BytesIO(uploaded['kidney_disease.csv']))
df.head()
```

```python
# Remove uneeded id column
df.drop('id', axis = 1, inplace = True)
```

```python
df.head()
```

```python
# View data info
df.info()     # Some are numbers some are objects
```

```python
# Some of the data are objects, although they are numeric in dataset. Convert them to numeric
df['pcv'] = pd.to_numeric(df['pcv'], errors = 'coerce')
df['wc'] = pd.to_numeric(df['wc'], errors = 'coerce')
df['rc'] = pd.to_numeric(df['rc'], errors = 'coerce')
```

```python
df.info()
```

```python
# Separate numeric and categorical (object) data
num_columns = [col for col in df.columns if df[col].dtype != 'object']
cat_columns = [col for col in df.columns if df[col].dtype == 'object']
```

```python
num_columns
```

```python
cat_columns
```

```python
# Check available values of each category column
for col in cat_columns:
    print(f"{col} has {df[col].unique()} values\n")
```

```python
# Some columns need to be fixed (ex: '\tno' needs to be replaced with 'no')
df['dm'].replace(to_replace = {'\tno' :'no' ,'\tyes' :'yes' , ' yes' :'yes'}, inplace = True)
df['cad'].replace(to_replace = '\tno' , value = 'no', inplace = True)
df['classification'].replace(to_replace = "ckd\t" , value =" ckd", inplace = True) 
```

```python
for col in cat_columns:
    print(f"{col} has {df[col].unique()} values\n")
```

```python
# Some columns have null values, therefore we will imputate these values (populate)
df.isna().sum()
```

```python
df[num_columns].isnull().sum()
```

```python
df[cat_columns].isnull().sum()
```

```python
# Random sampling for higher null values and mean/mode sampling for lower null values
def random_sample(feature):
    random_sample = df[feature].dropna().sample(df[feature].isna().sum())
    random_sample.index = df[df[feature].isnull()].index
    df.loc[df[feature].isnull(),feature] = random_sample
    
def impute_method(feature):
    frequent_feature = df[feature].mode()[0]
    df[feature] = df[feature].fillna(frequent_feature)
```

```python
# Fill numeric columns using random sampling
for col in num_columns:
    random_sample(col)
```

```python
df[num_columns].isnull().sum()
```

```python
# Random sampling for rbc and pc (greater null values)
random_sample('rbc')
random_sample('pc')

# mean/mode sampling for lower null values
for col in cat_columns:
    impute_method(col)
```

```python
df[cat_columns].isnull().sum()
```

```python
# Convert categorical data into labels of 1 and 0
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for col in cat_columns:
  df[col] = encoder.fit_transform(df[col])
```

```python
df.head()
```

```python
# Split data
X = df.drop(columns = 'classification', axis = 1)
Y = df['classification']
```

```python
X
```

```python
# feature selection - take only columns that have an effect
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

feture_el_model  = SelectFromModel(Lasso(alpha=0.005 ,random_state =42))

feture_el_model.fit(X, Y)
feture_el_model.get_support()
```

```python
cols = X.columns
selected_feature = cols[feture_el_model.get_support()]
```

```python
Xg = X[selected_feature]
```

```python
Xg
```

```python
# Split data into train and test
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(Xg, Y, test_size = .3, random_state = 42)
```

```python
# Create multiple models to obtain best model

from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

models = []
models.append(('LogisticRegression',LogisticRegression(solver='liblinear')))
models.append(('Naive bayes' ,GaussianNB()))
models.append(('Random Forest' ,RandomForestClassifier()))
models.append(('Decision Tree' ,DecisionTreeClassifier()))
models.append (('KNN' ,KNeighborsClassifier()))
```

```python
for name, model in models:
    model.fit(x_train, y_train)
    prediction = model.predict(x_test)
    print(name)
    print(confusion_matrix(prediction, y_test))
    print(accuracy_score(prediction,y_test))
    print('\n')
```

```python
model = models[2][1]
prediction = model.predict(x_test)   
print('Random Forest')
print(confusion_matrix(prediction, y_test))
print(accuracy_score(prediction, y_test))
print('\n')
```
